### Getting Surnames

```python
def finding_surnames(name):

    surnames = []
    paternal = False
    maternal = False
    elements = name.split(' ')
    target = ''

    for element in elements:
        
        if not target:
            target += element
        else:
            target += ' ' + element
        
        response = df_paternos_total.index.str.contains(target).any()
        
        if response and not paternal:
            paternal = True
            surnames.append(target)
            target = ''
        elif response and not maternal:
            maternal = True
            surnames.append(target)
            target = ''
    
    if not surnames:
        #surnames = ['NOT FOUND', 'NOT FOUND']
        surnames = [elements[0], elements[1]]
    if len(surnames) == 1:
        surnames.append( elements[1] )
    return surnames 

df_planta['Apellido Paterno'] = df_planta.apply(lambda L: finding_surnames(L.nombre)[0], axis=1)
```
